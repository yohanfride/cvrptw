```python
import numpy as np
import pandas as pd
import json
import random
import warnings
warnings.filterwarnings('ignore')


# get customer demand and location data
n = 50
customer = [i for i in range(1, n + 1)]
node = [0] + customer
arcos = [(i,j) for i in node for j in node if i != j]
f = open('pelanggan4.json')
data = json.load(f)
df = pd.DataFrame(data)
distance = np.loadtxt('distance.txt')
vehicle_capacity = 250
demand = df.demand
duration = np.loadtxt('duration.txt')
# Service time
ser = [df['service_time'][i] for i in range(n + 1)]
ser[0] = 0
# Start time
e = [df['ready_time'][i] for i in range(n + 1)]
e[0] = 0
# Due time
l = [df['due_time'][i] for i in range(n + 1)]
```

```python
def serviceTimeChek(s_time,i,j):
    cekData =  data[j]
    cekTime = s_time + duration[i][j]
    print(str(i)+"-"+str(j)+" --> "+str(cekData["ready_time"])+" - "+str(cekData["due_time"])+" --> "+str(cekTime))
    if cekTime >= cekData["ready_time"] and cekTime <= cekData["due_time"] :
        return True
    else:
        return False

def serviceTimeChek2(s_time,i,j):
    cekData =  data[j]
    cekTime = s_time + duration[i][j]
    print(str(i)+"-"+str(j)+" --> "+str(cekData["ready_time"])+" - "+str(cekData["due_time"])+" --> "+str(cekTime))
    if cekTime <= cekData["due_time"] :
        return True
    else:
        return False

def saving(data_distance,data_duration,node,i,s_time):
    if i == 0:
        [i,j] = savingFirst(data_distance, data_duration, node, s_time)
        res = i
        return res
    max_saving = 0
    max_saving2 = 0
    res = 0
    res2 = 0
    for j in range(1,n+1):
        if (j in node) and (j != i):
            S1 = data_distance[i][0] + data_distance[0][j] - data_distance[i][j]
            if(max_saving<S1):
                if(serviceTimeChek(s_time,i,j)):
                    max_saving = S1
                    res = j
            if(max_saving2<S1):
                if(serviceTimeChek2(s_time,i,j)):
                    max_saving2 = S1
                    res2 = j
    if(res == 0):
        res = res2
    return res

def savingFirst(data_distance, data_duration, node, s_time):
    max_saving = 0
    res = [0,0]
    max_saving2= 0
    res2= [0,0]
    for i in range(1,n+1):
        for j in range(1,n+1):
            if (j in node) and (i in node) and (j != i):
                S1 = data_distance[i][0] + data_distance[0][j] - data_distance[i][j]
                if(max_saving<S1):
                    if(serviceTimeChek(s_time,i,j)):
                        max_saving = S1
                        res = [i,j]
                if(max_saving2<S1):
                    if(serviceTimeChek2(s_time,i,j)):
                        max_saving2 = S1
                        res2 = [i,j]
                    
    if(res == [0,0]):
        res = res2
    return res

def action(nodes, routes, capacity, s_time, i, j):
    routes.append(j)
    s_time+=duration[i][j]
    if(j != 0):
        nodes.remove(j)
        capacity-=df.demand[j]
        s_time+=ser[j]
    i = j    
    return nodes, routes, capacity, s_time, i
```

```python
nodes = node[1:]
print(nodes)
route = 1
routes = [0]
croute = []
capacity = vehicle_capacity

###Menghitung Waktu
import time
start_timer = time.time()
i = 0
s_time = 0
while(len(nodes)>0):
    j = saving(distance,duration,nodes,i,s_time)
    if (capacity - demand[j] > 0) and (j > 0) :
        nodes,routes,capacity,s_time,i = action(nodes,routes,capacity,s_time,i,j)
    else:
        nodes,routes,capacity,s_time,i = action(nodes,routes,capacity,s_time,i,0)
        croute.append(routes)
        route += 1
        routes = [0]
        capacity = vehicle_capacity
        s_time = 0
        
    if(len(nodes) == 0):
        nodes,routes,capacity,s_time,i = action(nodes,routes,capacity,s_time,i,0)
        croute.append(routes)    
    print(routes)
    
###Menghitung Waktu
done_timer = time.time()
elapsed = done_timer - start_timer
```

```python
totals = 0
for h in range(len(croute)):
    print("Rute "+str(h+1))
    total = 0
    permintaan = 0
    for m in range(1,len(croute[h])) :
        total+=distance[croute[h][m-1]][croute[h][m]]
        permintaan+=df.demand[croute[h][m]]
    print(str(croute[h])+" -> "+str(total)+" -> "+str(permintaan))
    totals+=total 
print(totals)
print(elapsed)
```
