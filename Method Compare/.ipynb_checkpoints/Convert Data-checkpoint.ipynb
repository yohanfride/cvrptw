```python
import json, random
f = open('pelanggan.json')
pelanggan = json.load(f)
TimeProcess = 240
TimeRange = 1200
# random 10 - 3000
pelanggan[0]['ready_time'] = 0
pelanggan[0]['due_time'] = 5000
pelanggan[0]['id'] = 0
epsilon = 0.1
for j in range(1, len(pelanggan)):
    pelanggan[j]['service_time'] = TimeProcess
    greed = random.random()
    if(greed<=epsilon):
        start = random.randint(1, 300)  
    else:
        start = random.randint(100, 1200)
    pelanggan[j]['ready_time'] = start
    pelanggan[j]['due_time'] = start + TimeRange
    pelanggan[j]['id'] = j
    pelanggan[j]['best_arrive'] = 0

with open('pelanggan4.json', 'w') as outfile:
    json.dump(pelanggan, outfile)
```

```python
import json, random
import numpy as np
import pandas as pd
import gmaps
import googlemaps

API_KEY = 'AIzaSyCDq2h9qLZ1zKRHMT5Qv2QBfGWL0sAKDzI'
gmaps.configure(api_key=API_KEY)
googlemaps = googlemaps.Client(key=API_KEY)

f = open('pelanggan.json')
data = json.load(f)

# make dataframe which contains vending machine location and demand
df = pd.DataFrame(data)
print(df)



# function for calculating distance between two pins
def _duration_calculator(_df):
    
    _duration_result = np.zeros((len(_df),len(_df)))
    _df['latitude-longitude'] = '0'
    for i in range(len(_df)):
        _df['latitude-longitude'].iloc[i] = str(_df.latitude[i]) + ',' + str(_df.longitude[i])
    
    for i in range(len(_df)):
        for j in range(len(_df)):
            
            # calculate distance of all pairs
            _google_maps_api_result = googlemaps.directions(_df['latitude-longitude'].iloc[i],
                                                            _df['latitude-longitude'].iloc[j],
                                                            mode = 'driving')
            # append distance to result list
            _duration_result[i][j] = _google_maps_api_result[0]['legs'][0]['duration']['value']
    
    return _duration_result

duration = _duration_calculator(df)
print(duration)
```

```python
np.savetxt('duration.txt', duration)
```
